```python
import pandas as pd
```

```python
cj = pd.read_csv('cj_category.csv')
cj
```

```python
cj.drop('Unnamed: 0', axis=1, inplace=True)
cj
```

```python
cj.drop_duplicates()
```

```python
cj.info()
```

```python
cj = cj.astype({'category': 'string'})
```

```python
cj.info()
```

```python
cj
```

```python
cj = cj.groupby('item').agg({'category': lambda x: ",".join(set(x))})
```
