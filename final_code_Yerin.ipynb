## 모델링

### 임직원



1. Gradient Boosting

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ShineHop/BDA_CJTheMarket/cj_cate2_sungyeon/cj_cate2.csv')
df
```

```python
df_emp_y = df.query('employee_yn == "Y"')
df_emp_y
```

```python
df_emp_y = df_emp_y.drop('employee_yn', axis=1)
df_emp_y
```

```python
df_emp_y.set_index('scd',inplace=True)
df_emp_y
```

```python
# X, y 나누기
X = df_emp_y.drop('prime_yn', axis=1)
y = df_emp_y['prime_yn']
```

```python
# train, valid 데이터셋 나누기
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify=y, random_state=42)
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

# 모델 학습, 파라미터 조정
gradient_boosting = GradientBoostingClassifier()
gradient_boosting.fit(X_train, y_train)

# 예측
pred = gradient_boosting.predict(X_valid)

# 정확도
accuracy = accuracy_score(y_valid, pred)
print("Accuracy:", accuracy)

# 평가
cfreport = classification_report(y_valid, pred)
print("분류예측 레포트:\n", cfreport)

f1 = f1_score(y_valid, pred)
print("f1 score:\n", f1)
```

- 파라미터 조정 안한게 가장 성능 높음

2. XGBoost

```python
# XGBoost
import xgboost as xgb

# 모델 학습, 파라미터 조정
xgb = xgb.XGBClassifier(learning_rate=0.01, max_depth=5, n_estimators=1000)
xgb.fit(X_train, y_train)

# 예측
pred = xgb.predict(X_valid)

# 정확도
accuracy = accuracy_score(y_valid, pred)
print("Accuracy:", accuracy)

# 평가
cfreport = classification_report(y_valid, pred)
print("분류예측 레포트:\n", cfreport)

f1 = f1_score(y_valid, pred)
print("f1 score:\n", f1)
```

- Gradient Boosting 모델이 더 성능 좋음 : f1-score = 0.77
- XGBoost : f1-score = 0.76
