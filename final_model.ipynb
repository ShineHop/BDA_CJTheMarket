## 모델링

### 임직원



1. Gradient Boosting

```python
import pandas as pd
df_emp_y = pd.read_csv('전처리 완료된 임직원 데이터 깃허브 주소')
df_emp_y
```

```python
# X, y 나누기
X = df_emp_y.drop('prime_yn', axis=1)
y = df_emp_y['prime_yn']
```

```python
# train, valid 데이터셋 나누기
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, stratify=y, random_state=42)
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

# 모델 학습, 파라미터 조정
gradient_boosting = GradientBoostingClassifier('learning_rate': 0.01, 'max_depth': 5, 'n_estimators': 1000)
gradient_boosting.fit(X_train, y_train)

# 예측
pred = gradient_boosting.predict(X_valid)

# 정확도
accuracy = accuracy_score(y_valid, pred)
print("Accuracy:", accuracy)

# 평가
cfreport = classification_report(y_valid, pred)
print("분류예측 레포트:\n", cfreport)

f1 = f1_score(y_valid, pred)
print("f1 score:\n", f1)
```

2. XGBoost

```python
# XGBoost
import xgboost as xgb

# 모델 학습, 파라미터 조정
xgb = xgb.XGBClassifier('learning_rate': 0.01, 'max_depth': 5, 'n_estimators': 1000)
xgb.fit(X_train, y_train)

# 예측
pred = xgb.predict(X_valid)

# 정확도
accuracy = accuracy_score(y_valid, pred)
print("Accuracy:", accuracy)

# 평가
cfreport = classification_report(y_valid, pred)
print("분류예측 레포트:\n", cfreport)

f1 = f1_score(y_valid, pred)
print("f1 score:\n", f1)
```

```python
test_data.info()
```

```python
# prime_yn, Unnamed: 0 컬럼 삭제
test_data.drop('Unnamed: 0', axis=1, inplace=True)
test_data.drop('prime_yn', axis=1, inplace=True)
test_data
```

```python
# 인코딩 수행

# 성별 인코딩 (1 = Male, 0 = Female)
map_gender = {'F':0, 'M':1}
test_data['gender'] = test_data['gender'].map(map_gender)

test_data
```

```python
# 임직원 여부 인코딩 (1 = 임직원, 0 = 일반)
map_employee = {'N':0, 'Y':1}
test_data['employee_yn'] = test_data['employee_yn'].map(map_employee)
test_data
```

```python
# 구매일자 인코딩 (1 = 평일, 0 = 주말 및 공휴일)
test_data['weekday'] = test_data['order_date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
test_data
```

```python
# 1 = 평일, 0 = 주말 및 공휴일(설 연휴)
import numpy as np
test_data['weekday'] = test_data['order_date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
test_data['weekday'] = np.where((test_data['weekday'].dt.weekday==5) | (test_data['weekday'].dt.weekday==6) | (test_data['weekday']==2023-1-23) | (test_data['weekday']==2023-1-24), 0, 1)
test_data
```

```python
# order_date 컬럼 삭제
test_data.drop('order_date', axis=1, inplace=True) 
test_data
```
