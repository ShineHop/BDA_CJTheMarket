```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tmk_bda_train.csv')
df
```

```python
df.columns
```

```python
df.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
# 프라임 회원 유무 인코딩 (1 = 프라임 회원, 0 = 일반 회원)
map_prime = {'N':0, 'Y':1}
df['prime_yn'] = df['prime_yn'].map(map_prime)
df
```

```python
# 성별 인코딩 (1 = Male, 0 = Female)
map_gender = {'F':0, 'M':1}
df['gender'] = df['gender'].map(map_gender)
df
```

```python
# 구매일자 인코딩
df['weekday'] = df['order_date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
df.head()
```

```python
df['weekday'] = df['weekday'].dt.weekday
df
```

```python
# 요일로 분류 (0 = 월, 1 = 화, 2 = 수, 3 = 목, 4 = 금, 5 = 토, 6 = 일)
df['weekday'].value_counts()
```

```python
# 1월 23일 월, 1월 24일 화 : 설 연휴
df['weekday2'] = df['order_date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
df['weekday2'] = np.where((df['weekday2'].dt.weekday==5) | (df['weekday2'].dt.weekday==6) | (df['weekday2']==2023-1-23) | (df['weekday2']==2023-1-24), 0, 1)
df
```

```python
# 평일 / 주말 및 공휴일로 분류 (1 = 평일, 0 = 주말 및 공휴일)
df['weekday2'].value_counts()
```

```python
# 이벤트 할인 제품 총 구매 횟수 (event_product)
df['event_product'] = df['product_name'].apply(lambda x: x if x[0]=='[' else 0)
```

## 주문번호별 합치기

```python
# order_date은 weekday 와 weekday2로 바꿔줬으니 지우기
df.drop('order_date',axis=1,inplace=True)
df
```

```python
# 행 몇개가 합쳐졌는지 확인하기 위함
# 나중에 쓸 일이 생김
df['row_num'] = 1
df
```

```python
# 행을 기준으로 다 합쳐주기 이 과정에서 object인 productname을 제외하고 다 더해짐
df_group = df.groupby('scd').sum()
df_group
```

```python
# net_order_qty, net_order_amt, event_product는 더해진채로 둔다
# 각각 총 주문 수량, 총 주문 금액, 총 주문한 이벤트 제품 수량이 된다.
```

```python
# 나머지 값들은 다 같은 값이나 행들만큼 더해진 상황이다
# 따라서 전에 만든 row_num만큼 나누어주면 된다.
df_group['gender'] = df_group['gender'] / df_group['row_num']
df_group['age_grp'] = df_group['age_grp'] / df_group['row_num']
df_group['prime_yn'] = df_group['prime_yn'] / df_group['row_num']
df_group['weekday'] = df_group['weekday'] / df_group['row_num']
df_group['weekday2'] = df_group['weekday2'] / df_group['row_num']
df_group
```

```python
df_group.info()
```

```python
# 데이터 타입을 int로 바꿔주기
df_group[['gender','age_grp','prime_yn','weekday','weekday2']] = df_group[['gender','age_grp','prime_yn','weekday','weekday2']].astype(int)
df_group
```
